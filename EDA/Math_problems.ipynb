```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generate a parabola function

```python
x=np.arange(-10,10,0.1)
y=x**2
fig,ax=plt.subplots(figsize=(10,10))
ax.plot(x,y)
ax.set_xlabel("X")
ax.set_ylabel("X squared");
```

```python
x=np.arange(-10,10,0.1)
y=x**3
fig,ax=plt.subplots(figsize=(10,10))
ax.plot(x,y)
ax.set_xlabel("X")
ax.set_ylabel("X squared");
```
